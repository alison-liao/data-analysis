```python
#列出超10%缺失值的有序特征
#concat函数 axis： 需要合并链接的轴，0是行，1是列 
total = df.isnull().sum().sort_values(ascending=False) 
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False) 
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Missing Percent']) 
missing_data['Missing Percent'] = missing_data['Missing Percent'].apply(lambda x: x * 100) #替换成百分数
missing_data.loc[missing_data['Missing Percent'] > 10] #loc函数选取行和列,根据这个index来索引对应的行。
#df.loc[0, :]，这样可以清楚的看出为第0行的所有记录，同样如果取第’A’列的所有记录，可以写df.loc[:, ‘A’]
```
